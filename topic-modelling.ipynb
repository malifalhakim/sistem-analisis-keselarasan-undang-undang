* Import Libraries

```python
import pandas as pd
from bertopic import BERTopic
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from umap import UMAP
from sentence_transformers import SentenceTransformer
from bertopic.representation import KeyBERTInspired
from bertopic.representation import MaximalMarginalRelevance
from bertopic.representation import TextGeneration
import random
```

## Load Data

```python
df = pd.read_csv('/kaggle/input/undang-undang/full-dataset-after-processing.csv')
```

```python
df.tail()
```

```python
docs = df['BUNYI_AYAT']
```

## Default Representation Model

```python
# Prevent Stochastic Behavior (Mencegah Hasil Berubah-ubah setiap Menjalankan)
umap_model = UMAP(random_state=10)
```

* Without removing the stopwords

1. Modelling and Training

```python
topic_model = BERTopic(umap_model=umap_model,language="indonesian",min_topic_size=2,verbose=True)
topics, probs = topic_model.fit_transform(docs)
```

2. Analysis

```python
topic_model.get_topic_info()
```

```python
topic_model.get_document_info(docs)
```

```python
topic_base_df = topic_model.get_topic_info()
topic_base_doc_df = topic_model.get_document_info(docs)
```

```python
number_of_topics = topic_base_df['Topic'].max()
no_topic = random.randint(0,number_of_topics)
topic_x_df = topic_base_doc_df[topic_base_doc_df['Topic'] == no_topic].reset_index(drop=True)
```

```python
print('Topic :',no_topic)

print('AYAT 1')
print(topic_x_df.loc[0,'Document'])

print('\nAYAT 2')
print(topic_x_df.loc[1,'Document'])
```

3. Save topic to dataset for backup

```python
baseline = pd.concat([df,topic_base_doc_df[['Topic']]],axis=1)
baseline.head()
```

```python
baseline.to_csv('baseline-topic.csv', index=False)
```

4. Vizualise Top 8 Topic

```python
topic_model.visualize_barchart(title='Topic pada Undang-Undang')
```

## Another Representation Model

1. Define Representasion Model

```python
representation_model_1 = KeyBERTInspired()
representation_model_2 = MaximalMarginalRelevance(diversity=0.3)

list_representation_model = [representation_model_1,representation_model_2]
```

2. Pre Calculate Embeddings to Speed Up Training

```python
# Pre-calculate embeddings
embedding_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
embeddings = embedding_model.encode(docs, show_progress_bar=True)
```

3. Preventing Stochastic Behavior

```python
umap_model = umap_model
```

## MMR Representation Model

1. Modelling

```python
representation_model = list_representation_model[1]
```

```python
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    language="indonesian", 
    min_topic_size=2, 
    representation_model=representation_model,
    verbose = True, 
)
```

2. Training

```python
topics, probs = topic_model.fit_transform(docs, embeddings)
```

3. Analysis

```python
topic_model.get_topic_info()
```

```python
topic_model.get_document_info(docs)
```

```python
topic_base_df = topic_model.get_topic_info()
topic_base_doc_df = topic_model.get_document_info(docs)
```

```python
number_of_topics = topic_base_df['Topic'].max()
no_topic = random.randint(0,number_of_topics)
topic_x_df = topic_base_doc_df[topic_base_doc_df['Topic'] == no_topic].reset_index(drop=True)
```

```python
print('Topic :',no_topic)

print('AYAT 1')
print(topic_x_df.loc[0,'Document'])

print('\nAYAT 2')
print(topic_x_df.loc[1,'Document'])
```

## KeyBERT with Topic Reduction

1. Modelling

```python
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    language="indonesian", 
    min_topic_size=2, 
    representation_model=representation_model,
    verbose = True, 
    nr_topics="auto"
)
```

2. Training

```python
topics, probs = topic_model.fit_transform(docs, embeddings)
```

3. Analysis

```python
topic_model.get_topic_info()
```

```python
topic_model.get_document_info(docs)
```

```python
topic_base_df = topic_model.get_topic_info()
topic_base_doc_df = topic_model.get_document_info(docs)
```

```python
number_of_topics = topic_base_df['Topic'].max()
no_topic = random.randint(0,number_of_topics)
topic_x_df = topic_base_doc_df[topic_base_doc_df['Topic'] == no_topic].reset_index(drop=True)
```

```python
print('Topic :',no_topic)

print('AYAT 1')
print(topic_x_df.loc[0,'Document'])

print('\nAYAT 2')
print(topic_x_df.loc[1,'Document'])
```

4. Save

```python
keybert_tr = pd.concat([df,topic_base_doc_df[['Topic']]],axis=1)
keybert_tr.head()
```

```python
keybert_tr.to_csv('keybert-tr-topic.csv', index=False)
```

5. Visualize

```python
topic_model.visualize_barchart(title='Topic pada Undang-Undang')
```

6. Outliers Reduction

```python
# Reduce outliers with pre-calculate embeddings instead
new_topics = topic_model.reduce_outliers(docs, topics, strategy="embeddings", embeddings=embeddings)
```

```python
import collections
frequency = collections.Counter(new_topics)
frequency = dict(frequency)
print(dict(sorted(frequency.items(), key=lambda item: item[1])))
```

```python
keybert_tr_or = keybert_tr.copy()
keybert_tr_or['Topic'] = new_topics
```
